Задания по NumPy

```python
import numpy as np
import pandas as pd

X = np.random.random_sample((100, 5))
X
```

```python
print(X.max())
print(X.min())
print(X.mean())
print(X.sum())
print(X.std())
print(np.median(X))

print('\n')
for i in range(5):
  print("max, mean, sum & std of column #", i, ":\n", X[:, i].max(), "\t", X[:, i].mean(), "\t", X[:, i].sum(), "\t",X[:, i].std())
  print()

print("--------------------------------------------")
print("Same but different approach:")
max = X.max(axis=0)
mean = X.mean(axis=0)
sum = X.sum(axis=0)
std = X.std(axis=0)
for i in range(5):
  print("max, mean, sum & std of column #", i, ":")
  print(max[i], "\t", mean[i], "\t", sum[i], "\t", std[i])
  print()
```

```python
w = np.random.randint(2, 9, 5)
print(w)

y = X @ w
print(y)
```

Задания по Pandas

```python
df = pd.DataFrame(X)
df.columns = ["Column #0", "Column #1", "Column #2", "Column #3", "Column #4"]
print(df)
```

```python
df["target"] = (df.mean(axis=1) >= 0.5).astype(int)
print(df)
```
